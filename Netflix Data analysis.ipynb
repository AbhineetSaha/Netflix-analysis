```python
#filling the NA values of director
df['director']=df['director'].fillna('Anonymous')
df.head()
```

```python
dirlist=pd.DataFrame()
dirlist=df['director']
dirlist=dirlist.to_frame()
dirlist.head(10)
```

```python
directors=dirlist.groupby(['director']).size().reset_index(name='Total_Count')
print(directors.head(10))
```

```python
directors=directors[directors.director!='Anonymous']
print(directors)
```

```python
directors=directors.sort_values(by=['Total_Count'],ascending=False)
print(directors.head(10))
```

```python
bar=px.bar(directors.head(10),x='Total_Count',y='director',title='Top 10 directors')
bar.show()
```

```python
df['cast']=df['cast'].fillna('Unknown')
castdf=pd.DataFrame()
castdf=df['cast']
castdf=castdf.to_frame()
castdf.columns=['Actor']
print(castdf.head(10))
```

```python
actors=castdf.groupby(['Actor']).size().reset_index(name='Total Count')
actors=actors[actors.Actor!='Unknown']
actors=actors.sort_values(by=['Total Count'],ascending=False)
print(actors.head(10))
```

```python
bar2=px.bar(actors.head(10),x='Total Count',y='Actor',title='Top 10 actors casted')
bar2.show()
```

```python
df3=df[['type','release_year']]
df4=df3.groupby(['release_year','type']).size().reset_index(name='Total Count')
df4=df4[df4['release_year']>2010]
print(df4)
```

Analysing the trend of movie and tv show per year


```python
graph=px.line(df4,x='release_year',y='Total Count',color='type',title='Trend over the years')
graph.show()
```

```python
#Polarity of the content (Sentimental Analysis)
df5=df[['release_year','description']]
for index,row in df5.iterrows():
  d=row['description']
  testimonial=TextBlob(d)
  p=testimonial.sentiment.polarity
  if p==0:
    sent='Neutral'
  elif p>0:
    sent='Positive'
  else:
    sent='Negative'
  df5.loc[index,'Sentiment']=sent # Use index directly to set the value for the 'Sentiment' column in the current row
df5=df5.groupby(['release_year','Sentiment']).size().reset_index(name='Total Count')
df5=df5[df5['release_year']>2010]
bar3=px.bar(df5,x='release_year',y='Total Count',color='Sentiment',title='Sentiment Analysis')
bar3.show()
```

```python
categorical_cols = ['type', 'director', 'cast']
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
```

```python
from sklearn.pipeline import Pipeline
categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))])
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ('cat', categorical_transformer, categorical_cols)])
```

```python
model = Pipeline(steps=[('preprocessor', preprocessor),
                        ('classifier', RandomForestClassifier(n_estimators=100))])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
print(classification_report(y_test, y_pred))
```
